### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
#Some of the number values are stored as strings, when they should be stored as numbers. 
#Use update_many to convert latitude and longitude to decimal numbers.

# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
from bson.decimal128 import Decimal128
from decimal import *

pFlavours =  establishments.find({'BusinessName': 'Penang Flavours'})[0]

pLat = pFlavours["geocode"]["latitude"].to_decimal()
pLong = pFlavours["geocode"]["longitude"].to_decimal()

distance = Decimal("0.01")

pLeftLat = pLat - distance
pRightLat = pLat + distance

pLeftLong = pLong - distance
pRightLong = pLong + distance

print ("++++++++++++++++")
print(pLat)
print(pLong)
print ("++++++++++++++++")


query = {
         "$and": [
            {'geocode.latitude': {'$gte': Decimal128(pLeftLat), '$lte': Decimal128(pRightLat)}},
            {'geocode.longitude': {'$gte': Decimal128(pLeftLong), '$lte': Decimal128(pRightLong)}},
            {'RatingValue': {'$eq':'5'} },
             
        ] }
    
# Sort by hygiene score    
sort = [('scores.Hygiene',1)] 

limit = 5
# Print the results
results=list(establishments.find(query).sort(sort).limit(limit))
pprint(results)
```

```python
# Convert result to Pandas DataFrame
result_df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(result_df))
# Display the first 10 rows of the DataFrame
result_df.head()
```
